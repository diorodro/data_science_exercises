```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
base_sinasc_ro = pd.read_csv('sinasc_RO_2019.csv')
base_sinasc_ro.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
sns.countplot(x='APGAR1', data=base_sinasc_ro)
```

```python
sns.countplot(x='APGAR5', data=base_sinasc_ro)
```

```python
sns.countplot(x='SEXO', data=base_sinasc_ro)
```

```python
sns.countplot(x='GESTACAO', data=base_sinasc_ro)
```

```python
sns.countplot(x='ESTCIVMAE', data=base_sinasc_ro)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='APGAR5',y = 'CONSULTAS', data=base_sinasc_ro)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(20, 10))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0,0], x='SEXO',data=base_sinasc_ro)
sns.countplot(ax = axes[0,1], x='GESTACAO',data=base_sinasc_ro)
sns.histplot(ax = axes[1,0], x='ESTCIVMAE',data=base_sinasc_ro)
sns.histplot(ax = axes[1,1], x='CONSULTAS',data=base_sinasc_ro)
```
